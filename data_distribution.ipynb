# Data Distribution

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as stata
```

```python
# Load Data
df = sns.load_dataset("tips")
df.head()
```

```python
# Draw a hist plot on KDE
sns.histplot(data=df,x='total_bill',kde=True)
plt.figure(figsize=(8,6))
```

**Insights**
- Data is right skewed.
- Data is discrete.

```python
sns.histplot(data=df,x='tip',kde=True)
```

**Insights**
- Distribution is not Normal
- Sometimes people give High or Low tip, it is not Uniform. It depends on People.
- Data is discrete.

```python
sns.histplot(data=df,x='size',kde=True)
```

**Insights**
- You can see the Distribution of table in people.
- Normal Distribution towards two person table.
- Data is discrete.
- Pair of 2 person come around 160

```python
# Load Titanic Dataset
df2 = sns.load_dataset("Titanic")
df2.head()
```

```python
sns.histplot(data=df2, x='age', kde = True)
```

**Insights**
- Data is normally Distributed, except some of kids in the start.

```python
sns.histplot(data=df2, x='fare', kde = True, bins = 20)
```

**Insights**
- Data is right skewed. This term as Gamma Distribution.

```python
sns.histplot(data=df2, x='age', kde = True, bins = 20, hue='who')
```

**Insights**
- This shows distribution of data curve affected due to children.

```python
sns.boxplot(data=df2, x='age', hue='who')
```
